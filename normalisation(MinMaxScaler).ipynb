```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("c:\\Users\\mitta\\OneDrive\\Desktop\\DATASET\\data_set\\covid_toy - covid_toy.csv")
df.head()
```

```python
df['fever'] = df['fever'].fillna(df['fever'].mean())
```

```python
df['gender'] = df['gender'].map({"Male" : 0 , 'Female' : 1})
df['has_covid'] = df['has_covid'].map({'Yes':1 , 'No':0})
df['cough'] = df['cough'].map( {"Mild" : 0 , 'Strong': 1 })
```

```python
df = df.drop(columns=['city'])
```

```python
x = df.drop(columns=['has_covid'])
y = df['has_covid']
 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state= 42)
```

```python
np.round(x_train.describe(),2)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn = MinMaxScaler()
```

```python
x_train_mn = mn.fit_transform(x_train)
```

```python
x_train_new = pd.DataFrame(x_train_mn , columns = x_train.columns )
```

```python
np.round(x_train_new.describe(),2)
```
